```python
from ultralytics import YOLO
import os
```

```python
DATASET_CONFIG_PATH = '../datasets/wildlife_dataset.yaml'
```

```python
PRETRAINED_MODEL = '../yolo12s.pt'  
EPOCHS = 100
IMAGE_SIZE = 640
BATCH_SIZE = 8
```

```python
model = YOLO(PRETRAINED_MODEL)
```

```python
print("Starting model training...")
results = model.train(
   data=DATASET_CONFIG_PATH,
   epochs=EPOCHS,
   imgsz=IMAGE_SIZE,
   batch=BATCH_SIZE
)

print("\nTraining complete!")
print(f"The final model and results are saved in: {results.save_dir}")
```
